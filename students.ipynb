```python
import os
import operator
############################
diccionarioEstudiantes = {}
############################
# Función para "limpiar" la consola
def clear(): 
    os.system('cls')
############################
#Función para adicionar un estudiante

def adicionarEstudiante(codigo, nombre, nota):
    global diccionarioEstudiantes

    clear()
    diccionarioEstudiantes[codigo] = [nombre,nota]

    print("!Estudiante adicionado! Datos : {}".format(diccionarioEstudiantes[codigo]))
    
    input("Digite cualquier tecla para continuar...")

############################
#Funcion que imprime la sábana de notas.
def imprimirSabanaNotas():
    global diccionarioEstudiantes

    clear()
    print("  ---SÁBANA DE NOTAS---  ")
    print("")
    print("Código\t\tNombre\t\tNota")
    for clave in diccionarioEstudiantes:
        print(str(clave) + "\t\t" + diccionarioEstudiantes[clave][0] + "\t\t" + str(diccionarioEstudiantes[clave][1]))
    print("")
    print("---FIN DE LA SÁBANA DE NOTAS---")
    input("Digite cualquier tecla para continuar...")

############################
#Función que permite calcular la sumatoria de las notas
def calcularSumatoriaNotas():
    global diccionarioEstudiantes

    sum = 0
    for clave in diccionarioEstudiantes:
        sum = sum + diccionarioEstudiantes[clave][1]
    return sum

############################
#Función que permite llenar el diccionario con valores por defecto
def llenardiccionarioEstudiantesPrueba():
    global diccionarioEstudiantes

    diccionarioEstudiantes[123] = ["Carlos",4.5]
    diccionarioEstudiantes[456] = ["Andres",3.0]
    diccionarioEstudiantes[789] = ["Luis",5.0]
    diccionarioEstudiantes[111] = ["Paola",3.7]
    diccionarioEstudiantes[222] = ["Luisa",5.0]
    diccionarioEstudiantes[444] = ["Alejo",4.8]
    diccionarioEstudiantes[333] = ["Gaby",3.3]
    
############################
#Función que permite calcular el promedio de las notas
def calcularPromedioNotas():
    global diccionarioEstudiantes

    sum = calcularSumatoriaNotas()

    #for clave in diccionarioEstudiantes:
    #    sum = sum + diccionarioEstudiantes[clave][1]
    if len(diccionarioEstudiantes) == 0:
        raise NameError("División por 0")
    else:
        return sum / len(diccionarioEstudiantes)
############################
# Función que elimina un estudiante del diccionario
def eliminarEstudiante(codigo):
    global diccionarioEstudiantes
    try:
        del(diccionarioEstudiantes[codigo])
    except: 
        raise NameError("El estudiante de código " + str(codigo) + " No existe!")
############################
# Función que actualiza la nota de un estudiante del diccionario
def actualizarNotaEstudiante(codigo,nota):
    global diccionarioEstudiantes
    try:
        diccionarioEstudiantes[codigo][1] = nota
    except: 
        raise NameError("El estudiante " + str(codigo) + " No existe!")

############################
def imprimirInfoDelEstudiante(codigo):
    global diccionarioEstudiantes
    clear()
    try:
        print ("ESTUDIANTE : ")
        print ("Código : " + str(codigo))
        print ("Nombre : " + diccionarioEstudiantes[codigo][0])
        print ("Nota : " + str(diccionarioEstudiantes[codigo][1]))
        
    except: 
        raise NameError("El estudiante " + str(codigo) + " No existe!")

############################
#Función para encontrar el código del estudiante con la nota mas alta
def encontrarEstudianteNotaMasAlta():
    global diccionarioEstudiantes
    
    notaMayor = 0
    codigo = -1

    for clave in diccionarioEstudiantes:
        if(diccionarioEstudiantes[clave][1] >= notaMayor):
            notaMayor = diccionarioEstudiantes[clave][1]
            codigo = clave

    return codigo

############################
#Función para encontrar el código del estudiante con la nota mas baja
def encontrarEstudianteNotaMasBaja():
    global diccionarioEstudiantes
    
    notaMenor = float('inf')
    codigo = -1

    for clave in diccionarioEstudiantes:
        if(diccionarioEstudiantes[clave][1] <= notaMenor):
            notaMenor = diccionarioEstudiantes[clave][1]
            codigo = clave

    return codigo

############################
#Funcion para ordenar el diccionario por Nombre del estudiante
def ordenarDiccionarioEstudiantesPorNombre():
    global diccionarioEstudiantes
    diccionarioEstudiantes = dict(sorted(diccionarioEstudiantes.items(), key=lambda x: x[1][0]))

############################
#Funcion para ordenar el diccionario por Nota del estudiante
def ordenarDiccionarioEstudiantesPorNota():
    global diccionarioEstudiantes
    diccionarioEstudiantes = dict(sorted(diccionarioEstudiantes.items(), key=lambda x: x[1][1]))
    
############################
#Funcion para ordenar el diccionario por Nota del estudiante utilizando ciclo

def ordenarDiccionarioEstudiantesPorNotaCiclo():
    global diccionarioEstudiantes
    temporal = {}
    while(diccionarioEstudiantes):
        codigo = encontrarEstudianteNotaMasBaja()
        temporal[codigo] = [diccionarioEstudiantes[codigo][0],diccionarioEstudiantes[codigo][1]]
        diccionarioEstudiantes.pop(codigo)
    diccionarioEstudiantes = temporal        
############################
############################

llenardiccionarioEstudiantesPrueba()
while(True):
    clear()
    print("MENÚ GESTIÓN DE ESTUDIANTES")
    print("")
    print("(0) Salir")
    print("(1) Adicionar Estudiante")
    print("(2) Eliminar Estudiante")
    print("(3) Actualizar Estudiante")
    print("(4) Mostrar Estudiante")
    print("(5) Imprimir sábana de notas")
    print("(6) Calcular la sumatoria de las notas")
    print("(7) Calcular el promedio de las notas")
    print("(8) Encontrar estudiante con la Nota más alta")
    print("(9) Encontrar estudiante con la Nota más Baja")
    print("(10) Ordenar la lista de estudiantes por Nota")
    print("")
    opc = input("Digite una opción: ")
    opc = int(opc)

    if opc == 0:
        break
    elif opc == 1:
        clear()
        print("Ingrese los datos del estudiante")
        print("")
        cod = int(input("Digite código estudiante: "))
        nombre = input("Digite nombre estudiante: ")
        nota = float(input("Digite nota estudiante: "))
        adicionarEstudiante(cod,nombre, nota)

    elif opc == 2:
        clear()
        print("")
        cod = int(input("Ingrese el código del estudiante a eliminar : "))
        ###Confirmar!
        try:
            eliminarEstudiante(cod)
            print("El estudiante de código {} fue eliminado correctamente!".format(cod))
            input("Digite cualquier tecla para continuar...")
        except Exception as e:
            print("Error : " + str(e))
            input("Digite cualquier tecla para continuar...")
    
    elif opc == 3:
        clear()
        print("")
        try:
            cod = int(input("Ingrese el código del estudiante a Actualizar : "))
            imprimirInfoDelEstudiante(cod)
            print("")
            print("")
            nota= float(input("Ingrese la nota del estudiante a Actualizar : "))
            actualizarNotaEstudiante(cod,nota)
            print("El estudiante de código {} fue actualizado correctamente con la nota {:.2f}".format(cod,nota))
            input("Digite cualquier tecla para continuar...")
        except Exception as e:
            print("Error : " + str(e))
            input("Digite cualquier tecla para continuar...")

    
    elif opc == 4:
        clear()
        cod = int(input("Ingrese el código del estudiante a Actualizar : "))
        imprimirInfoDelEstudiante(cod)
        input("Digite cualquier tecla para continuar...")
    
    elif opc == 5:
        imprimirSabanaNotas()

    elif opc == 6:
        clear()
        sum = calcularSumatoriaNotas()
        print("La sumatoria de notas es : {:.2f}".format(sum))
        input("Digite cualquier tecla para continuar...")
    
    
    elif opc == 7:
        clear()
        try:
            sum = calcularPromedioNotas()
            print("El promedio de notas es : {:.2f}".format(sum))
            input("Digite cualquier tecla para continuar...")
        except Exception as e:
            print("Error : " + str(e))
            input("Digite cualquier tecla para continuar...")
    
    
    elif opc == 8:
        cod = encontrarEstudianteNotaMasAlta()
        clear()
        imprimirInfoDelEstudiante(cod)
        input("Digite cualquier tecla para continuar...")

    elif opc == 9:
        cod = encontrarEstudianteNotaMasBaja()
        clear()
        imprimirInfoDelEstudiante(cod)
        input("Digite cualquier tecla para continuar...")


    elif opc == 10:
        clear()
        #ordenarDiccionarioEstudiantesPorNota()
        ordenarDiccionarioEstudiantesPorNotaCiclo()
        print("Lista Ordenada!")
        input("Digite cualquier tecla para continuar...")
    
```
